```python
df = pd.read_csv('descriptors_output.csv')
df
```

```python
x = df.drop('Name', axis=1)
x
```

```python
y = df3['pIC50']
y
```

# Create Labeled Dataset

```python
dataset = pd.concat([x,y], axis=1)
dataset
```

```python
dataset.to_csv('acetylcholinesterase_06_bioactivity_data_3class_pIC50_pubchem_fp.csv', index=False)
```

# Remove Low Variance Features

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))    
x = selection.fit_transform(x)
```

```python
x.shape
```

# Split Dataset

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

# **Build a Regression Model using Random Forest**

```python
model1 = RandomForestRegressor(n_estimators=100)
model1.fit(x_train, y_train)
accuracy = model1.score(x_test, y_test)
accuracy
```

**Model Prediction**

```python
y_pred = model1.predict(x_test)
```

**Model Performance**

```python
mre = mean_squared_error(y_test, y_pred)
mre
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

# Build Model using LGBMRegressor

```python
import lightgbm as lgb
model2 = lgb.LGBMRegressor()
model2.fit(x_train, y_train)
accuracy = model2.score(x_test, y_test)
accuracy
```

**Model Prediction**

```python
y_pred2 = model2.predict(x_test)
```

**Model Performance**

```python
mre = mean_squared_error(y_test, y_pred2)
mre
```

```python
r2 = r2_score(y_test, y_pred2)
r2
```
