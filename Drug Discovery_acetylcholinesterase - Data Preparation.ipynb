```python
#import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

Search for Target Protein

```python
# Target search for acetylcholinesterase
target = new_client.target
target_query = target.search('acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

**Select and retrieve bioactivity data for Human Acetylcholinesterase (first entry)**

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

Retrieve only bioactivity data for Human Acetylcholinesterase (CHEMBL220) that are reported as IC50 values in nM (nanomolar) unit

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df
```

```python
df.to_csv('bioactivity_data_acetylcholinesterase.csv', index=False) 
```

**Handling missing data**

If any compounds has missing value for the standard_value column then drop it

```python
import pandas as pd
df = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/Bioinformatics/Acetylcholinesterase/bioactivity_data_acetylcholinesterase.csv')
```

```python
df2 = df[df.standard_value.notna()]
df2
```

```python
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

```python
df2 = df2.drop_duplicates(['canonical_smiles'])
df2
```

**Data pre-processing of the bioactivity data**

Labeling compounds as either being active, inactive of intermediate

The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be **active** while those greater than 10,000 nM will be considered to be **inactive**. As for those values in between 1,000 and 10,000 nM will be referred as **intermediate**.

```python
bioactivity_class = []
for i in df2.standard_value:
  if i >= 10000:
    bioactivity_class.append("inactive")
  elif i <= 1000:
    bioactivity_class.append("active")
  else: 
    bioactivity_class.append("intermediate")
```

**Combine the 3 lists into a dataframe**

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selection]
df3
```

```python
# pd.concat([df3,pd.Series(bioactivity_class)], names=['bioactivity_class'], axis=1)
df3.insert(3, 'bioactivity_class', pd.Series(bioactivity_class), True)
```

```python
df3 = df3[df3.bioactivity_class.notna()]
df3
```

```python
df3.to_csv('bioactivity_preprocessed_data_acetylcholinesterase.csv', index=False)
```
